# preprocessing

```python
df = pd.read_csv("../data/Espoo_Vantaa.csv")
df['postcode'] = df['postcode'].astype(str).str.pad(5, "left", "0")
df['price_m2'] = df['price'] / df['floor_area']

# 
df["municipality"]=df["municipality"].astype('category')
# number_of_room
df['number_of_room'] = df['house_info'].str.extract(pat =  "(\d\s*m*h*\s?\+*)")
df['number_of_room'] = df['number_of_room'].str.extract('(\d+)')
df['number_of_room'] = df['number_of_room'].map(lambda x: float(x) if not pd.isna(x) else x)

df['number_of_room_cat'] = df['number_of_room'].map(lambda x: str(round(x)) if not pd.isna(x) else None)
df['number_of_room_cat'] = df['number_of_room'].map(lambda x: "Over 3 rooms"if float(x) >3 else x)


# filter

df =df[ ~((df['number_of_room'] <1 )| (df['number_of_room'].isnull() )) ]
df =df[ df['price_m2']>1 ]
```

```python
(ggplot(df, aes('municipality', fill='property_type'))
 + geom_bar( position='dodge')+
theme_bw()+
ggtitle('Number of listings among different types of properties')+
xlab('Municipality') +
ylab("Listings"))    
```

```python
(ggplot(df, aes('property_type', "price_m2"))
 + geom_boxplot(fill='blue' )+
theme_bw()+
ggtitle('Asking price per m2 in Espoo and Vantaa')+
xlab('Asking price per m2') +
ylab("Property Type")+
facet_wrap('municipality',  nrow=2)+
scale_y_continuous(limits=(0,5000)) )    
```

```python
(ggplot(df, aes('number_of_room_cat', "price_m2"))
 + geom_boxplot(fill='pink' )+
theme_bw()+
ggtitle('Asking price per m2 between among number of rooms in a unit in Espoo and Vantaa')+
xlab('Asking price per m2') +
ylab("Number of rooms")+
facet_wrap('municipality',  nrow=2)+
scale_y_continuous(limits=(0,5000)) )    
```

```python
df1 = df.copy()
df1['price_m2'] = np.log(df1['price_m2'])
(ggplot(df1, aes('build_year', "price_m2"))
 + geom_point(fill='pink' )+
scale_x_continuous(limits=(1900,2020))+
ggtitle('Asking price per m2 (in log) based on build year')+
xlab('Asking price per m2 (log)') +
ylab("Build year"))    
```
